```python
#Reading the text line by line
lines = convo.readlines()
```

```python
#Create a list to view the text neatly.
list_convo = []
#Using the loop function we add the elements to the list.
for line in lines:
    if (line=="\n"):
        continue
    list_convo.append(line[:-2])
    #'continue' statement in looping will skip "/n" from each element of list.
```

```python
 # Create a set to find out the number of unique characters in the sample conversation.
Speakers = set()
 for i in list_convo:
        a = i.split(":")[0]
        Speakers.add(a)
```

```python

a = '.'
for x, y in enumerate(list_convo):
    list_convo[x] = y.replace(a, ' ')
    

b = ','
for x, y in enumerate(list_convo):
    list_convo[x] = y.replace(b, ' ')
    
c = '?'
for x, y in enumerate(list_convo):
    list_convo[x] = y.replace(c, ' ') 

d = '...'
for x, y in enumerate(list_convo):
    list_convo[x] = y.replace(a, ' ')
```

```python
# Create an empty list "dialouges" to make each dialouge as an element in the list.
dialouges = []

# Use loop function to add the dialouges to the list.
for i in list_convo:
    a = i.split(':')
    dialouges.append(a)
```

```python
dialouges
```

```python
# Create an empty dictionary to generate a key value pair of unique speakers
unique = {}

# Here we are using looping function to add keys and values to the dictionary.

for i, j in dialouges:
    if i in unique.keys():
        unique[i].append(j)
    else:
        unique[i]=[j]
# Here each unique speaker represents the key and the dialouges represents the value.   
```

```python
unique
```

```python
unique.keys()
```

```python
#Create a list of unique speakers and with the help of loop function add elements to this list.
speakers = []

for i in unique.keys():
    speakers.append(i)
```

```python
speakers
```

```python
#Creating list of words spoken by each unique speaker using loop function
words = []

for speaker in speakers:
    s = unique[speaker]
    words.append(s)
```

```python
words
```

```python
# Here we have used loop function to create separate text file for each unique speaker with their name containing unique words.
for speaker in speakers:
    file = open(f'{speaker}.txt', 'a+', encoding = 'utf-8')
    m = speakers.index(speaker)
    n = words[m]
    j = " ".join([str(item) for item in n])
    mno = j.split(" ")
    unique_words = []
    for word in mno:
        if word not in unique_words:
            unique_words.append(word)
 
    for z in unique_words:
        file.write(z + '\n')
    file.close()
```
